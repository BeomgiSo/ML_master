# 문제해결 프로세스 및 체크리스트 정의

## 머신러닝 문제해결 프로세스

1. 문제의 이해
    - 문제 이해(배경,목적,유형 등)
    - 평가지표파악

1. 탐색적 데이터 분석 (Explorartory Data Analysis,EDA)
    - 데이터 구조 탐색
    - 데이터 시각화(중요피처파악)

1. 데이터 전처리

1. 베이스라인 모델
    - (선택)피처 엔지니어링
    - 모델훈련 및 성능 검증
    - 결과 예측 및 제출

1. 성능개선
    - 피처엔지니어링
        - 이상치 제거, 결측값 처리, 데이터 인코딩, 피처 스케일링, 파생피처(변수)생성, 피처 제거 및 선택
    - 하이퍼파라미터 최적화
        - 하이퍼라미터 튜닝 : 값의 범위는 제각각이다. 그중 모델 성능을 가장 좋게 맏느는 값을 찾는것이다. 가능한 범위 내에서 최선의 하이퍼파라미터를 찾아야한다
        - 최적화기법으로는 그리디서치, 랜덤서치, 베이지안 최적화 등이 있다.
    - 성능 검증
        - 훈련모델의 일반화 성능 평가한다
        - 일반화 성능이란 : 훈련때 접하지 못한 새로운데이터에 대해 예측하는 성능을 일반화 성능 이라고한다.
    - 결과 예측 및 제출
    

- 머신러닝 문제해결 체크리스트

### EDA

    1. EDA
        - 파일별 용도 파악
        - 데이터 양(레코드 수, 피처 수, 전체 용량 등)
        - 피처 이해(이름,의미,데이터타입,결측값 개수, 고유값 개수, 실제값, 데이터 종류 등)
        - 훈련데이터와 테스트 데이터의 차이
        - 타깃값
    2. 데이터 시각화
        - 효과적인 시각화를 통해 피처 엔지니어링 
            - 수치형 데이터 시각화 : 히스토그램, 커널밀도추정, 분포도, 러그플롯
            - 범주형 데이터 시각화 : 막대그래프, 포인트플롯, 박스플롯, 바이올린플롯, 카운트플롯
            - 데이터 관계 시각화 : 히트맵, 라인플롯, 산점도, 산점도 + 회귀플롯

    3. 피처 파악
        - 추가할 피처, 제거할 피처, 피처별 인코딩 전략
        - 이상치 파악 : 추가/제거 피처 목록, 인코딩 전략, 이상치 처리 전략

### 베이스라인 모델

    1. 준비하기
        - 데이터 불러오기
        - 기본적인 피처 엔지니어링
        - 평가자료 계산 함수 준비
            결과물 : 데이터, 평가지표 계산 함수
        
    1. 모델 훈련
        - 모델 생성
        - 훈련
            결과물 : 훈련된 베이스라인 모델

    1. 성능 검증
        - 예측(검증데이터사용)
        - 평가
            결과물 : 예측 결과, 검증평가점수

    1. 예측및 결과 제출
        - 최종예측(테스트데이터사용)
        - 제출 파일 생성
        - 제출

### 피처 엔지니어링

    1. 피처 엔지니어링
        - 이상치 제거, 피처 스케일링, 결측값 처리, 피쳐명 한글화, 데이터 인코딩, 데이터 다운스케일링,
        - 데이터 타입 변경, 데이터 조합 생성, 파생 피처 생성, 필요없는 피처 제거, 시차 피처 생성(시계열데이터인경우)

    2. 모델훈련 with 하이퍼파라미터 최적화
        - 하이퍼파라미터 종류와 의미 파악
        - 선별
            - 최적화할 하이퍼파라미터
            - 값을 고정할 하이퍼파라미터
        - 값의 범위 설정
        - 최적화 기법 : (그리디서치, 베이지안 서치, OOF예측)
        - 모델 생성 및 훈련(최적화)
            결과물 : 최적의 하이퍼파라미터 훈련, 훈련된 모델
        
    3. 성능검증
        - 예측(검증 데이터 사용)
        - 성능평가
            - 결과물 : 예측결과, 검증 평가 점수

    4. 예측 및 결과 제출
        - 최종 예측(테스트데이터 사용)
        - 제출파일 생성
        - 제출
